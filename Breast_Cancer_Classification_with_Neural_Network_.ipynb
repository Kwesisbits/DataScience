```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.datasets
from sklearn.model_selection import train_test_split
```

```python
#loading data from sklearn
breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
```

```python
print(breast_cancer_dataset)
```

```python
# loading the data to a dataframe
data_frame = pd.DataFrame(breast_cancer_dataset.data, columns = breast_cancer_dataset.feature_names)
```

```python
# print first 5 rows of dataframe
data_frame.head()
```

```python
#adding 'target' column to the dataframe
data_frame['label'] = breast_cancer_dataset.target
```

```python
# print lst 5 rows of dataframe
data_frame.tail()
```

```python
#number of rows and columns in the dataset
data_frame.shape
```

```python
#getting relevant info about data
data_frame.info()
```

```python
#Checking for missing values
data_frame.isnull().sum()
```

```python
# stactical measures about the data
data_frame.describe()
```

```python
# checking the distribution of Target Variable
data_frame['label'].value_counts()
```

1 -> Benign (non-cancerous, does not spread,usually less dangerous)
           

 0 -> Malignant (cancerous, harmful, can spread to other parts of the body)

```python
data_frame.groupby('label').mean()
```

Separating the features and target

```python
X = data_frame.drop(columns= 'label', axis=1)
Y = data_frame['label']
```

```python
print(X)
```

```python
print(Y)
```

Splitting the data into training data and testing data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Standardize the data

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()

X_train_std = scaler.fit_transform(X_train)
X_test_std = scaler.transform(X_test)
```

**Building the Neural Network**

```python
# importing relevant libraries
import tensorflow as tf
tf.random.set_seed(3)
from tensorflow import keras
```

```python
#setting up layers of Neural Network
model = keras.Sequential([
    keras.layers.Flatten(input_shape=(30,)),
    keras.layers.Dense(20, activation= 'relu'),
    keras.layers.Dense(2, activation= 'sigmoid')
])
```

```python
# compiling the Neural Network
model.compile(optimizer='adam',
              loss = 'sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
# training the neural network
history = model.fit(X_train_std, Y_train, validation_split=0.1, epochs=10)
```

Visualizing Accuracy and Loss

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')

plt.legend(['training data', 'validation data'], loc= 'lower right')
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')

plt.legend(['training data', 'validation data'], loc= 'upper right')
```

Accuracy of model on test data

```python
loss, accuracy = model.evaluate(X_test_std, Y_test)
print(accuracy)
```

```python
print(X_test_std.shape)
print(X_test_std[0])
```

```python
Y_pred = model.predict(X_test_std)
```

```python
print(Y_pred.shape)
print(Y_pred[0])
```

Model.predict gives prediction probablity of each class for data point

```python
#Converting prediction probability to class labels
Y_pred_labels = [int(np.argmax(i)) for i in Y_pred]
print(Y_pred_labels)
```

Building predictive system

```python
input_data = (17.99,10.38,122.8,1001,0.1184,0.2776,0.3001,0.1471,0.2419,0.07871,1.095,0.9053,8.589,153.4,0.006399,0.04904,0.05373,0.01587,0.03003,0.006193,25.38,17.33,184.6,2019,0.1622,0.6656,0.7119,0.2654,0.4601,0.1189)

#change input_data to a numpy array
input_data_numpy_array = np.asarray(input_data)

#reshape the numpy array as we are predicting for one data point
input_data_reshaped = input_data_numpy_array.reshape(1,-1)

#standardize input_data
input_data_std = scaler.transform(input_data_reshaped)

prediction = model.predict(input_data_std)

print(prediction)

prediction_label = [int(np.argmax(prediction))]
print(prediction_label)

if (prediction_label[0] == 0):
  print('The tumor is Malignant')

else:
    print('The tumor is Benign')

```
